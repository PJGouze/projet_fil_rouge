# Analyse des corrélation

## Importations

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Importations des données

```python
# !pip install openpyxl

fichier = "iodaa_2025_fichier_test.xlsx"
data = pd.read_excel(fichier)

# Affichage
display(data.head())
display(data.info())

# Sélection des variables numériques

variables_numeriques = data.select_dtypes(include=[np.number])

print("Variables numériques détectées :")
display(variables_numeriques.head())
```

## Matrice de corrélation globale

```python
correlations = variables_numeriques.corr(method='pearson')

plt.figure(figsize=(16,12))
sns.heatmap(
    correlations,
    annot=False,           # mettre True si vous voulez les valeurs
    cmap="coolwarm",
    center=0,
    square=True,
    cbar_kws={"shrink": 0.7}
)

plt.title("Matrice de corrélation - Variables numériques (Globale)", fontsize=16)
plt.show()
```

## Matrice avec en cible EM porc croissance uniquement

```python
cible = "EM porc croissance (kcal) kcal/kg brut"

# On enlève les autres cibles pour ne garder que les variables chimiques + la cible choisie
colonnes_a_retirer = [
    "ED porc croissance (kcal) kcal/kg brut",
    "EN porc croissance (kcal) kcal/kg brut",
    "EMAn coq (kcal) kcal/kg brut",
    "EMAn poulet (kcal) kcal/kg brut",
    "UFL 2018 par kg brut",
    "PDI 2018 g/kg brut",
    "BalProR u 2018 g/kg brut"
]

# Filtre
variables_filtrees = variables_numeriques.drop(columns=colonnes_a_retirer, errors='ignore')

# Matrice de corrélation filtrée
correlations_filtrees = variables_filtrees.corr()

plt.figure(figsize=(16,12))
sns.heatmap(
    correlations_filtrees,
    cmap="coolwarm",
    center=0,
    annot=False,
    square=True
)

plt.title("Matrice de corrélation — Variables chimiques + EM porc croissance", fontsize=16)
plt.show()
```

## Graphe corrélations des variables chimiques avec EM porc croissance

```python
corr_em = variables_filtrees.corr()[cible].sort_values(ascending=False)

plt.figure(figsize=(8,10))
sns.barplot(x=corr_em.values, y=corr_em.index)
plt.title("Corrélation des variables chimiques avec EM porc croissance")
plt.xlabel("Corrélation (Pearson)")
plt.ylabel("Variables")
plt.show()
```

## Analyse avec les différentes "cibles"

```python
# variables énergétiques
vars_energie = [
    "EB (kcal) kcal/kg brut", "ED porc croissance (kcal) kcal/kg brut",
    "EM porc croissance (kcal) kcal/kg brut", "EN porc croissance (kcal) kcal/kg brut",
    "EMAn coq (kcal) kcal/kg brut", "EMAn poulet (kcal) kcal/kg brut",
    "UFL 2018 par kg brut"
]

# corrélation chimie vs énergie
corr_subset = data[vars_compo + vars_energie].corr().loc[vars_compo, vars_energie]

# heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr_subset, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Corrélation entre compositions chimiques et énergies digestibles")
plt.xlabel("Énergies digestibles")
plt.ylabel("Compositions chimiques")
plt.show()
```

## Matrices de corrélations par classe (nom de céréale)

```python
# noms uniques de sous-classes
noms_uniques = data["Nom"].unique()

# liste des cibles à retirer (on garde seulement EM porc)
colonnes_a_retirer = [
    "ED porc croissance (kcal) kcal/kg brut",
    "EN porc croissance (kcal) kcal/kg brut",
    "EMAn coq (kcal) kcal/kg brut",
    "EMAn poulet (kcal) kcal/kg brut",
    "UFL 2018 par kg brut",
    "PDI 2018 g/kg brut",
    "BalProRu 2018 g/kg brut"
]

for sous_classe in noms_uniques:
    
    # données filtrées pour la sous-classe
    sous_data = data[data["Nom"] == sous_classe]
    print(f"Sous-classe : {sous_classe} ({len(sous_data)} lignes)")
    
    # sélection des variables numériques
    sous_num = sous_data.select_dtypes(include=[np.number])
    
    # filtre : on enlève les autres cibles
    sous_num = sous_num.drop(columns=colonnes_a_retirer, errors='ignore')
    
    # contrôle : au moins deux colonnes
    if sous_num.shape[1] < 2:
        print(f"Pas assez de variables numériques pour {sous_classe}")
        continue
    
    # matrice de corrélation
    corr = sous_num.corr()
    
    # affichage heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, cmap="coolwarm", center=0, square=True)
    plt.title(f"Matrice de corrélation – Sous-classe : {sous_classe}", fontsize=14)
    plt.show()
```

# Analyse en composantes principales

```python
# 1. VARIABLES QUALITATIVES / QUANTITATIVES

vars_qual = ["Classe", "Nom"]
vars_quant = [col for col in data.columns if col not in vars_qual]

# conversion en numérique
for col in vars_quant:
    data[col] = pd.to_numeric(data[col], errors="coerce")

# 2. NORMALISATION

from sklearn.preprocessing import StandardScaler
normaliseur = StandardScaler()
X_normalise = normaliseur.fit_transform(data[vars_quant])

# 3. ACP

from sklearn.decomposition import PCA
acp = PCA()
X_acp = acp.fit_transform(X_normalise)

acp_df = pd.DataFrame(X_acp, columns=[f"CP{i+1}" for i in range(X_acp.shape[1])])
acp_df["Classe"] = data["Classe"]
acp_df["Nom"] = data["Nom"]

# 4. SCREE PLOT

plt.figure(figsize=(7,4))
plt.plot(range(1, len(acp.explained_variance_ratio_)+1),
         acp.explained_variance_ratio_, marker='o')
plt.xlabel("Composante principale")
plt.ylabel("Variance expliquée (%)")
plt.title("Scree plot")
plt.grid(True)
plt.show()

# 5. CERCLE DE CORRÉLATION

composantes = acp.components_[0:2]

plt.figure(figsize=(6,6))
plt.axhline(0, color='grey')
plt.axvline(0, color='grey')

cercle = plt.Circle((0,0), 1, color='grey', fill=False, linestyle='--')
plt.gca().add_artist(cercle)

# variables chimiques
vars_compo = ["MS % brut", "PB % brut", "CB % brut", "MGR % brut", "MM % brut",
              "NDF % brut", "ADF % brut", "Lignine % brut", "Amidon % brut", "Sucres % brut"]

# variables énergétiques
vars_energie = ["EB (kcal) kcal/kg brut", "ED porc croissance (kcal) kcal/kg brut",
                "EM porc croissance (kcal) kcal/kg brut", "EN porc croissance (kcal) kcal/kg brut",
                "EMAn coq (kcal) kcal/kg brut", "EMAn poulet (kcal) kcal/kg brut",
                "UFL 2018 par kg brut", "PDI 2018 g/kg brut", "BalProRu 2018 g/kg brut"]

# légende
from matplotlib.patches import Patch
legende = [
    Patch(facecolor='tomato', edgecolor='tomato', label='Composition chimique'),
    Patch(facecolor='royalblue', edgecolor='royalblue', label='Énergies / digestibilités')
]

# flèches
for i, var in enumerate(vars_quant):
    if var in vars_compo:
        couleur = "tomato"
    elif var in vars_energie:
        couleur = "royalblue"
    else:
        couleur = "grey"
    plt.arrow(0, 0, composantes[0, i], composantes[1, i],
              head_width=0.03, head_length=0.03, linewidth=1.5, color=couleur)
    plt.text(composantes[0, i]*1.1, composantes[1, i]*1.1, var, fontsize=8, color=couleur)

plt.legend(handles=legende, loc="lower right")
plt.xlabel("CP1")
plt.ylabel("CP2")
plt.title("Cercle de corrélation (CP1/CP2)")
plt.axis("equal")
plt.show()

# 6. BIPLOT INDIVIDUS + VARIABLES

plt.figure(figsize=(12, 9))

palette_chaude = ["#E64B35", "#D0604D", "#F39B7F", "#FFA74F"]
palette_froide = ["#4DBBD5", "#00A087", "#3C5488", "#4A7EBB", "#2E8B92"]

noms_cereales = acp_df[acp_df["Classe"] == "Céréales"]["Nom"].unique()
noms_tourteaux = acp_df[acp_df["Classe"] == "Tourteaux d'oléagineux"]["Nom"].unique()

couleurs_cereales = {nom: palette_chaude[i % len(palette_chaude)]
                     for i, nom in enumerate(noms_cereales)}
couleurs_tourteaux = {nom: palette_froide[i % len(palette_froide)]
                      for i, nom in enumerate(noms_tourteaux)}

couleur_map = {**couleurs_cereales, **couleurs_tourteaux}

# points individuels
for nom in couleur_map:
    masque = acp_df["Nom"] == nom
    plt.scatter(
        acp_df.loc[masque, "CP1"],
        acp_df.loc[masque, "CP2"],
        color=couleur_map[nom],
        label=nom,
        alpha=0.85,
        s=60
    )

plt.title("ACP — Individus (Chaud = Céréales | Froid = Tourteaux d'oléagineux)")
plt.xlabel(f"CP1 ({acp.explained_variance_ratio_[0]*100:.1f}%)")
plt.ylabel(f"CP2 ({acp.explained_variance_ratio_[1]*100:.1f}%)")
plt.legend(title="Nom des produits", loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Projection sur 3 axes

```python
import plotly.graph_objects as go

# graphique 3D ACP
fig = go.Figure()

for nom in couleur_map:
    masque = acp_df["Nom"] == nom

    fig.add_trace(go.Scatter3d(
        x=acp_df.loc[masque, "CP1"],
        y=acp_df.loc[masque, "CP2"],
        z=acp_df.loc[masque, "CP3"],
        mode="markers",
        name=nom,
        marker=dict(
            size=5,
            color=couleur_map[nom],
            opacity=0.8
        ),
        hovertext=acp_df.loc[masque, "Nom"] + " (" + acp_df.loc[masque, "Classe"] + ")"
    ))

fig.update_layout(
    title="ACP 3D — Individus (Chaud = Céréales | Froid = Tourteaux d'oléagineux)",
    scene=dict(
        xaxis_title=f"CP1 ({acp.explained_variance_ratio_[0]*100:.1f}%)",
        yaxis_title=f"CP2 ({acp.explained_variance_ratio_[1]*100:.1f}%)",
        zaxis_title=f"CP3 ({acp.explained_variance_ratio_[2]*100:.1f}%)"
    ),
    legend_title="Nom des produits",
    width=900,
    height=700
)

fig.show()
```
